```python
# Buatlah program untuk menghitung BMI!

berat = int(input('Berat Badan (kg) : '))
tinggi = int(input('Tinggi Badan (cm) : '))

tinggi_m = tinggi / 100

nilai_bmi = berat / (tinggi_m ** 2)

print('Nilai BMI anda : ', nilai_bmi)

if nilai_bmi < 18.5:
 print('Underweight')
elif nilai_bmi >= 18.5 and nilai_bmi < 25:
 print('Normal')
elif nilai_bmi >= 25 and nilai_bmi < 30:
 print('Overweight')
elif nilai_bmi >= 30:
 print('Obesity')
```
